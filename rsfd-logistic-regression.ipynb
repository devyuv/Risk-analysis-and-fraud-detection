### Random Over Sampling


```python
import imblearn
from imblearn.over_sampling import RandomOverSampler
from collections import Counter
```

```python
ros = RandomOverSampler(random_state=0)
x_resample, y_resample = ros.fit_resample(x, y)
print((Counter(y_resample)))
```

```python
Counter(x_resample.shape)
```

```python
Counter(y_resample)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_resample,y_resample,test_size=0.2,random_state=110)
```

```python
Counter(y_train)
```

```python
Counter(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_sample_pred=model.predict(x_test)
y_sample_pred
```

```python
from sklearn.metrics import precision_recall_curve, auc, f1_score
precision, recall, thresholds = precision_recall_curve(y_test, y_sample_pred)
auc_pr = auc(recall, precision)
auc_pr
```

```python
f1 = f1_score(y_test, y_sample_pred)
f1
```
